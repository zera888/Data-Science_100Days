```python
import numpy as np
import pandas as pd
index = pd.date_range('1/1/2019', periods=20, freq='D')
series = pd.Series(range(20), index=index)
print('原始以日為單位取得時間序列如下:', '\n', series , '\n')

```

```python
#1. 將所有轉為周資料
ts1 = series.to_period(freq="w")
print('將原始時間資料轉為以周為單位取得時間序列如下:', '\n', ts1 , '\n')
```

```python
#2. 將周資料的值取平均
ts2 = series.resample('W').mean()
print('將以周為單位取得平均值如下:', '\n', ts2 , '\n')
```
