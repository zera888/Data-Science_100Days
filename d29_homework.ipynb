### Q2:計算標準常態分配，大於1，小於 -1 的機率有多大?
![](Q2.png)


```python
# Python實現常態分配
#1.設定使用參數
mu=0
sigma=1
x=np.linspace(-1, 1, 100)
# 2.產生 pdf圖
y = stats.norm.pdf(x,mu, sigma)
plt.plot(x, y, "b-", linewidth=2)
#plt.grid(True)
plt.title('pdf of Norma(-1,1)')
plt.show()

cumsum_probs =  stats.norm.cdf(x,mu, sigma)
plt.ylabel('P(X<=x)')
plt.xlabel('x')
plt.title(' cdf of Nomal(-1,1)')
plt.plot(x, cumsum_probs,"b-", linewidth=2)
plt.show()
print(cumsum_probs[1])
```

### Q3:X~N(2,4),x 服從常態分配，平均數為2,變異數為 4，計算 X小於 6 的機率有多大?

```python
## Q3
#  提示: 和 Q1,Q2 不同，要先標準化。
mu=2
sigma=2
x=np.linspace(0, 6, 100)
# 2.產生 pdf圖
y = stats.norm.pdf(x,mu, sigma)
plt.plot(x, y, "b-", linewidth=2)
#plt.grid(True)
plt.title('pdf of Norma(0,6)')
plt.show()

cumsum_probs =  stats.norm.cdf(x,mu, sigma)
plt.ylabel('P(X<=x)')
plt.xlabel('x')
plt.title(' cdf of Nomal(2,6)')
plt.plot(x, cumsum_probs,"b-", linewidth=2)
plt.show()
print(cumsum_probs[6])
```
