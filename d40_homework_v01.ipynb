```python
# library
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
import math
import statistics
import seaborn as sns
from IPython.display import display
import sklearn
# print(sklearn.__version__)
#如果只有 0.19 記得要更新至 最新版本
%matplotlib inline
import pingouin as pg
import researchpy

#讀入資料
pd=pd.read_csv('Titanic_train.csv ')

#產生新的一組資料型態
df=pd[['Survived','Age','Sex']]


# 刪除有遺失值的列
df.dropna(inplace=True)
print('顯示刪除有遺失列之後的資料')
display(df)
print()


#計算總共有幾個列
num_rows = df.shape[0]
print('nun of rows is:', num_rows)
```

### Q1: 產生一個新的變數(Age_above65_) Age>=65為 'Y'，其餘為'N'。


```python
#產生Age_above65的新欄位
def Age_map(x):
    if (x>=65):
        return ('1')
    else :
        return ('0')
df['Age_above65']=df['Age'].apply(Age_map)
display(df)

print('確認Age_above65的人數')
df[['Age_above65','Sex']].groupby('Sex').count()
```

### Q2: 添加女性和男性，產生一個新的變數(Age_above65_female)，女性或Age>=65為'Y'，其餘為'N'。
* 暗示: 觀看下面影片找答案，https://www.youtube.com/watch?v=X2d-wUt5azY

```python
#產生變數Age_above65_female

def func(a, b):
    if 'female' in a and b >= 65:
        return 1
    else:
        return 0
df['Age_above65_female'] = df.apply(lambda x: func(x.Sex, x.Age), axis = 1)
display(df)

print('確認Age_above65的女性人數')
df[['Age_above65_female','Sex']].groupby('Sex').count()
```
