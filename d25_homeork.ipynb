```python
#  當 "projection" 參數為“ortho"時，所得圖位地球儀截面
```

```python
# 導入開發套件
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import mpl_toolkits.basemap
print(mpl_toolkits.basemap.supported_projections)

# 新建地圖
map = Basemap(projection = 'ortho', lat_0 = 0, lon_0 = 0)
#’ortho’指正射投影，後面两個参數是设置中心點

# 给整個地圖上藍色
map.drawmapboundary(fill_color = 'lightblue')

# 给陸地塗上珊瑚色，湖泊塗上藍色
map.fillcontinents(color = 'lightgreen', lake_color = 'blue')

# 畫圖
map.drawcoastlines()

# 顯示结果
plt.show()
```

```python
#  當 "projection" 參數為“Mill"時，所得平展位面
```

```python
# 導入開發套件
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import mpl_toolkits.basemap
print(mpl_toolkits.basemap.supported_projections)

# 新建地圖
map = Basemap(projection = 'mill', lat_0 = 0, lon_0 = 0)
#’ortho’指正射投影，後面两個参數是设置中心點

# 给整個地圖上藍色
map.drawmapboundary(fill_color = 'lightblue')

# 给陸地塗上珊瑚色，湖泊塗上藍色
map.fillcontinents(color = 'green', lake_color = 'aqua')

# 畫圖
map.drawcoastlines()

# 顯示结果
plt.show()
```

```python
2.  查看 resolution / 經緯度座標的繪圖精細度
```

```python
# 新建地圖
map = Basemap(projection='merc',urcrnrlat=80, llcrnrlat=-80,llcrnrlon=-180, urcrnrlon=180)  #Basemap有很多屬性，這里全都使用默認参數

# 畫圖
map.drawcoastlines()

map.drawparallels(range(-90,91,30), color='#CCCCCC')
map.drawmeridians(range(-180,181,60), color='#CCCCCC')

#畫出州。
map.drawstates(color='b') 

# 畫出城市
map.drawcountries(color='darkred')

# 劃出海的顏色
map.drawmapboundary(fill_color='lightblue')

# 劃出陸地的顏色
map.fillcontinents(color = 'lightgreen')

# 顯示结果
plt.show()
```
