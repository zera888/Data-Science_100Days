```python
#  當 "projection" 參數為“ortho"時，所得圖位地球儀截面
```

```python
# 導入開發套件
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import numpy as np

#設定圖形大小
plt.figure(figsize=(5,5))

# 新建地圖
map = Basemap(projection = 'ortho', lat_0 = 0, lon_0 = 0)
#’ortho’指正射投影，後面两個参數是设置中心點

# 给整個地圖上藍色
map.drawmapboundary(fill_color = 'lightblue')

# 给陸地塗上珊瑚色，湖泊塗上藍色
map.fillcontinents(color = 'lightgreen', lake_color = 'blue')

# 畫圖
map.drawcoastlines()

# 顯示结果
plt.show()
```

```python
# 導入開發套件
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import mpl_toolkits.basemap
print(mpl_toolkits.basemap.supported_projections)

#設定圖形大小
plt.figure(figsize=(5,5))

# 新建地圖
map = Basemap(projection = 'mill', lat_0 = 0, lon_0 = 0)
#’ortho’指正射投影，後面两個参數是设置中心點

# 给整個地圖上藍色
map.drawmapboundary(fill_color = 'lightblue')

# 给陸地塗上珊瑚色，湖泊塗上藍色
map.fillcontinents(color = 'green', lake_color = 'aqua')

# 畫圖
map.drawcoastlines()

# 顯示结果
plt.show()
```

```python
# 新建地圖
map = Basemap(projection='merc',urcrnrlat=80, llcrnrlat=-80,llcrnrlon=-180, urcrnrlon=180)  #Basemap有很多屬性，這里全都使用默認参數

# 畫圖
map.drawcoastlines()

map.drawparallels(range(-90,91,30), color='#CCCCCC')
map.drawmeridians(range(-180,181,60), color='#CCCCCC')

#畫出州。
map.drawstates(color='b') 

# 畫出城市
map.drawcountries(color='darkred')

# 劃出海的顏色
map.drawmapboundary(fill_color='lightblue')

# 劃出陸地的顏色
map.fillcontinents(color = 'lightgreen')

# 顯示结果
plt.show()
```

```python

import numpy as np
from itertools import chain

#運用 "itertools" 庫中的 "chain" 模組，可以繪製更接近於生活中所見世界地圖的圖像
fig = plt.figure(figsize=(8, 6), edgecolor='w')
m=Basemap(projection='moll',resolution=None,lat_0=0,lon_0=0)


# 繪製一個陰影浮雕的地理圖像
m.shadedrelief(scale=0.2)

# 經度與緯度將會以字典的形式返回
lats = m.drawparallels(np.linspace(-90, 90, 13))
lons = m.drawmeridians(np.linspace(-180, 180, 13))

# 運用 plt.Line2D 實例設置經緯線
lat_lines = chain(*(tup[1][0] for tup in lats.items()))
lon_lines = chain(*(tup[1][0] for tup in lons.items()))
all_lines = chain(lat_lines, lon_lines)

# 通過迴圈設置所有線條的樣式
for line in all_lines:
    print(type(line))
    line.set(linestyle='-', alpha=0.3, color='w')
plt.show()
```
