```python
# 題目 : 將資料夾中boston.csv讀進來，並用圖表分析欄位。
# 1.畫出箱型圖，並判斷哪個欄位的中位數在300~400之間?
# 2.畫出散佈圖 x='NOX', y='DIS' ，並說明這兩欄位有什麼關係?

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd = pd.read_csv('boston.csv')

#1.畫出箱型圖，並判斷哪個欄位的中位數在300~400之間?

#箱型圖
pd.boxplot()
plt.tight_layout()
plt.subplots_adjust() # 方法更改 Matplotlib 子圖間距 
plt.savefig('d12_fig')
plt.show()

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#2. 畫出散佈圖 x='NOX', y='DIS' ，並說明這兩欄位有什麼關係?
#散佈圖
df = pd.read_csv('boston.csv')
plt.scatter(df['NOX'],df['DIS'])
plt.gca().set(xlim=(0.01,1.0), ylim=(0, 12),
              xlabel='Polution', ylabel='die people')

plt.xticks(fontsize=12); plt.yticks(fontsize=12)
plt.title("NOX pulotion vs Dies people", fontsize=22)
plt.legend(fontsize=12)    
plt.show()


```
