### 1. 根據題目給的 DataFrame 完成下列操作：

```
- 計算每個不同種類 animal 的 age 的平均數
```

````python
# 記得先 Import 正確的套件

import numpy as np
import pandas as pd
````

````python
data = {
    'animal': ['cat', 'cat', 'snake', 'dog', 'dog', 'cat', 'snake', 'cat', 'dog', 'dog'],
    'age': [2.5, 3, 0.5, np.nan, 5, 2, 4.5, np.nan, 7, 3],
    'visits': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
    'priority': ['yes', 'yes', 'no', 'yes', 'no', 'no', 'no', 'yes', 'no', 'no']
}

labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
df = pd.DataFrame(data, index=labels)
df
````

````python
# 計算每個不同種類 animal 的 age 的平均數
cat = df[df.animal == 'cat']
print('Cat Age Avg:', cat.age.mean(axis=0))

dog = df[df.animal == 'dog']
print('Dog Age Avg:', dog.age.mean(axis=0))

snake = df[df.animal == 'snake']
print('Snake Age Avg:', cat.age.mean(axis=0))
````

````python
# 將資料依照 Age 欄位由小到大排序
df.sort_values('age', ascending=False)
````

````python
# 再依照 visits 欄位由大到小排序
df.sort_values('visits', ascending=False)
````

````python
# 將 priority 欄位中的 yes 和 no 字串，換成是布林值 的 True 和 False
df1 =df
df1['priority'] = df1['priority'].str.replace('yes', 'True')
df1['priority'] = df1['priority'].str.replace('no', 'False')
df1
````

````python
import pandas as pd
score_df = pd.DataFrame([[1,56,66,70], 
              [2,90,45,34],
              [3,45,32,55],
              [4,70,77,89],
              [5,56,80,70],
              [6,60,54,55],
              [7,45,70,79],
              [8,34,77,76],
              [9,25,87,60],
              [10,88,40,43]],columns=['student_id','math_score','english_score','chinese_score'])
score_df = score_df.set_index('student_id')
score_df
````

````python
#  6 號學生(student_id=6) 3 科平均分數為何
id6_mean = score_df[5:6].mean(axis=1).mean()
id6_mean
````

````python
# 6 號學生 3 科平均分數是否有贏過班上一半的同學？
overall_median = score_df.mean(axis=1).median()
if id6_mean > overall_median:
  print('Yes');
else: 
  print('No');
````

````python
#  由於班上同學成績不好，所以學校統一加分，加分方式為開根號乘以十，請問 6 號同學 3 科成績分別是？
score_plus = score_df.apply(lambda x : x**(0.5)*10)
score_plus[5:6]   
    
````

````python
# 承上題，加分後各科班平均變多少？
score_plus.mean()
````
