```python
import matplotlib.pyplot as plt
import numpy as np 

#---畫正弦cos曲線--- 

x = np.arange(0.3*np.pi,2*np.pi,0.01) 
a=5
b=4
y = np.cos(a*x+b) #畫線 plt.plot(x,y,color='green') 
plt.plot(x,y)
plt.xlabel('x values')
plt.ylabel('y values')
plt.title('cos function') #標題 
plt.grid(True) #圖例，注意中括號
plt.savefig("filename.png",dpi=300,format="png")
```
